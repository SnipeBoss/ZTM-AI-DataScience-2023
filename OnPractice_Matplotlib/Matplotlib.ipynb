## Two options for supplots

### Supplot option1

```python
# Subplot obtion 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10,5))
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10),
            np.random.random(10))
ax3.hist(np.random.randn(1000))
ax4.bar(dt_nut.keys(), dt_nut.values())
```

```python
# Subplot obtion 1
fig, ((ax1, ax2)) = plt.subplots(nrows = 2,
                                 ncols = 1,
                                 figsize = (10,5))
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(x, x/2)
```

## Which one should you use? pyplot vs matplotlib OO method?
#### When plotting something quick PYPLOT but if it advance use OO method

```python
dtf_heartdis_over50 = dtf_heartdis[dtf_heartdis["age"] > 50]
dtf_heartdis_over50
```

### Pyplot method

```python
dtf_heartdis_over50.plot(kind = 'scatter',
                         x = 'age',
                         y = 'chol',
                         c = 'target')
```

### OO method mixed with pyplot method

```python
fix, ax = plt.subplots(figsize = (10,6))
dtf_heartdis_over50.plot(kind = 'scatter',
                         x = 'age',
                         y = 'chol',
                         c = 'target',
                         ax = ax)
# to adjust x axis
ax.set_xlim([45, 90])
```

### OO method from scracth

```python
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = dtf_heartdis_over50["age"],
                     y = dtf_heartdis_over50["chol"],
                     c = dtf_heartdis_over50["target"])
# Customize the plot
ax.set(title="Heart disease and Cholesterol",
       xlabel = "age",
       ylabel = "Chol")

# Add a legend -> grab the value from scatter and unpacked
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line
# specifying horizontal line type 
# plt.axhline(y = 300, color = 'r', linestyle = '-')
# specifying horizontal line type 
plt.axhline(y = dtf_heartdis_over50["chol"].mean(), color = 'r', linestyle = '--') 
```

#### Subplot of chol, age, thelach

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Add data ax0
scatter = ax0.scatter(x = dtf_heartdis_over50['age'],
              y = dtf_heartdis_over50['chol'],
              c = dtf_heartdis_over50['target'])

#Customize ax0
ax0.set(title = "Heart disease and chol lev",
        xlabel = "Age",
        ylabel = "Chol")

# Add a legend ax0
ax0.legend(*scatter.legend_elements(), title = "Target")
# Add a mean line
ax0.axhline(y = dtf_heartdis_over50["chol"].mean(), color = 'r', linestyle = '--')

# Add data ax1
scatter = ax1.scatter(x = dtf_heartdis_over50['age'],
              y = dtf_heartdis_over50['thalach'],
              c = dtf_heartdis_over50['target'])

#Customize ax1
ax1.set(title = "Heart disease and thalach lev",
        xlabel = "Age",
        ylabel = "thalach")

# Add a legend ax1
ax1.legend(*scatter.legend_elements(), title = "Target")
# Add a mean line
ax1.axhline(y = dtf_heartdis_over50["thalach"].mean(), color = 'r', linestyle = '-') 

# Add a title
fig.suptitle("Heart DIS ANA", fontsize = 16, fontweight = "bold")
```
